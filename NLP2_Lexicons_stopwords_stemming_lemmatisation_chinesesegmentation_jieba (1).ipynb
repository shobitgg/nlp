### TASK 3: Stemming Paragraph

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
example = "The lazy dogs were jumped over by those cruel and cunning wolfs"
example = [stemmer.stem(token) for token in example.split(" ")]
print(" ".join(example))
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
example = "Chilla Christ his family name rhymes with mist was a left-arm spin bowler who also bowled medium-pace later in his career. He was first chosen to play for Queensland against Victoria in 1930-31 when they were short of players, but the match was rained off, and he had to wait another seven years for his debut"
example = [stemmer.stem(token) for token in example.split(" ")]
print(" ".join(example))
```

### TASK 4: Lemmatizer
