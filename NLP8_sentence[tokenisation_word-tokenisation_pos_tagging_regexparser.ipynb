### sentence tokenisation_word tokenisation _pos tagging _ regex parser

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk import pos_tag
```

```python
data = "Chamunda also known as Chamundeshwari, Chamundi, and Raktandika is a fearsome form of Chandi, the Hindu Divine Mother Parvati and is one of the seven Matrikas (mother goddesses). She is adyashakti and parashakti.She is also one of the chief Yoginis, a group of sixty-four or eighty-one Tantric goddesses, who are attendants of the warrior goddess parvati.[3] The name is a combination of Chanda and Munda, two monsters whom Chamunda killed. She is closely associated with Kali, another fierce aspect of Parvati. She is identified with goddesses Parvati, Kali or Durga.The goddess is often portrayed as haunting cremation grounds or fig trees. The goddess is worshipped by ritual animal sacrifices along with offerings of wine and in ancient times, human sacrifices were offered too. The practice of animal sacrifices has become less common with Shaivite and Vaishnavite influences"
```

### #Sentence Tokenize

```python
#Sentence Tokenize
sen_token = sent_tokenize(data)
sen_token
```

### word tokenization

```python
# word tokenization
word_token = word_tokenize(data)
word_token
```

### POS tagging

```python
# POS tagging
pos = nltk.pos_tag(word_token)
pos
```

### RegexpParser

```python
sentence = pos
grammar = "VP: {<MD>?<VB.*><NP|PP>}" 
cp = nltk.RegexpParser(grammar)
result = cp.parse(sentence)
print(result)
```
